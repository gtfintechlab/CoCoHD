```python
import pandas as pd
import numpy as np
```

Here's the schema for all energy data:
- DATE (datetime), PRICE, VOLUME, RETURN1/5/10/21 (float)

```python
crude_oil_df = pd.read_csv('/Users/mingxiaosong/Desktop/congress_nlp/energy_data/raw/Crude_Oil_Price_WTI.csv')
crude_oil_df['DATE'] = pd.to_datetime(crude_oil_df['DATE'])
crude_oil_df = crude_oil_df.rename(columns={'DCOILWTICO':'PRICE'})
crude_oil_df['PRICE'] = pd.to_numeric(crude_oil_df['PRICE'], errors='coerce')
crude_oil_df['VOLUME'] = np.nan
crude_oil_df['RETURN_1'] = crude_oil_df['PRICE'].pct_change(1).shift(-1)
crude_oil_df['RETURN_5'] = crude_oil_df['PRICE'].pct_change(5).shift(-5)
crude_oil_df['RETURN_10'] = crude_oil_df['PRICE'].pct_change(10).shift(-10)
crude_oil_df['RETURN_21'] = crude_oil_df['PRICE'].pct_change(21).shift(-21)
# crude_oil_df.to_csv('crude_oil.csv')
crude_oil_df
```
